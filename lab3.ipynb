```python
class Person:

    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        
    def __str__(self):
        return "\nFirst name = " + self.first_name + "\nLast name = " + self.last_name + \
                "\nAge = " + str(self.Age)
    
    
class Employee(Person):
    
    def __init__(self, first_name, last_name, employee_number, position, salary):
        
        self.first_name = first_name
        self.last_name = last_name
        self.employee_number = employee_number
        self.position = position
        self.salary = salary

    def raise_amt(self, amt):
        self.salary += amt
        
    def change_position(self, pos):
        self.position = pos
        
    def __str__(self):
        
        return "\nFirst name = " + self.first_name + "\nLast name = " + self.last_name + \
                "\nEmployee Number = " + str(self.employee_number) + "\nPosition = " + self.position + \
                "\nSalary = " + str(self.salary)
    

class Student(Person):
    
    def __init__(self, first_name, last_name, student_number, current_courses):
        
        self.first_name = first_name
        self.last_name = last_name
        self.student_number = student_number
        self.current_courses = current_courses
        
    def drop_course(self, course):
        self.current_courses.remove(course)
        
    def add_course(self, course):
        self.current_courses.append(course)
    
    def __str__(self):
        
        return "\nFirst name = " + self.first_name + "\nLast name = " + self.last_name + \
                "\nStudent Number = " + self.student_number + "\nCurrent Courses = " + self.current_courses 
               
    
    
    
    
    
    
    
```
